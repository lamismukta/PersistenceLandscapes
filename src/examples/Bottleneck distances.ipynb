```python
import gudhi as gd
import numpy as np
import pandas as pd
```

```python
import sys
sys.path.append('../functions')
from barcodes import get_data, gudhi_barcode, bottleneck, bottleneck_by_dimension
```

```python
data_3kzn = np.genfromtxt("../data/3kzn_chain_A.xyz", delimiter=' ')
data_4jkj = np.genfromtxt("../data/4jkj_chain_A.xyz", delimiter=' ')
```

```python
barcode_3kzn = gudhi_barcode(data_3kzn)
gd.plot_persistence_diagram(barcode_3kzn)
gd.plot_persistence_barcode(barcode_3kzn)
```

Second chain

```python
barcode_4jkj = gudhi_barcode(data_4jkj)
gd.plot_persistence_diagram(barcode_4jkj)
gd.plot_persistence_barcode(barcode_4jkj)
```

Bottleneck distance between 3kzn and 4jkj

```python
d = bottleneck(barcode_3kzn, barcode_4jkj)
print('Bottleneck distance between 3kzn and 4jkj is',d)
```

What if you split the homology classes?

```python
d_sep = bottleneck_by_dimension(barcode_3kzn, barcode_4jkj)
print('Bottleneck distance between 3kzn and 4jkj, when homology classes are seperated is', d_sep)
```

Case when they are different:

```python
data_6yma = get_data('6yma','A')
data_6ymb = get_data('6ymb','A')
barcode_6yma = gudhi_barcode(data_6yma)
barcode_6ymb = gudhi_barcode(data_6ymb)

gd.plot_persistence_diagram(barcode_6yma)
gd.plot_persistence_diagram(barcode_6ymb)
```

```python
print("Bottleneck distance", bottleneck(barcode_6yma, barcode_6ymb))
print("Bottleneck distance, split by dimension", bottleneck_by_dimension(barcode_6yma, barcode_6ymb))
```
