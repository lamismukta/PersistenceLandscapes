```python
import numpy as np
import time
from ripser import Rips
import matplotlib.pyplot as plt
```

```python
import sys
sys.path.append('../functions')
from barcodes import ripser_plot, interpolate
```

```python
data_3kzn = np.genfromtxt("../data/3kzn_chain_A.xyz", delimiter=' ')
tic = time.time()
ripser_plot(data_3kzn, title = 'Persistence Diagram for 3kzn chain')
toc = time.time()
times = {0: toc - tic}
```

Interpolating Points

```python
test_itp = interpolate(data_3kzn,1)
print('original data length is', len(data_3kzn))
print('interpolated data length is', len(test_itp))
```

Now I'll try it with Ripser

```python
one_extra_point = interpolate(data_3kzn,1)
tic = time.time()
ripser_plot(one_extra_point, title = 'Persistence Diagram for 3kzn chain with one extra point')
toc = time.time()
times[1] = toc - tic
```

```python
two_extra_points = interpolate(data_3kzn,2)
tic = time.time()
ripser_plot(two_extra_points, title = 'Persistence Diagram for 3kzn chain with three extra points')
toc = time.time()
times[2] = toc - tic
```

More extra points...

```python
three_extra_points = interpolate(data_3kzn,3)
tic = time.time()
ripser_plot(three_extra_points, title = 'Persistence Diagram for 3kzn chain with three extra points')
toc = time.time()
times[3] = toc - tic
```

```python
sorted_times = dict(sorted(times.items()))
x_values = list(sorted_times.keys())
y_values = list(sorted_times.values())
plt.plot(x_values, y_values)
plt.xlabel("No. points")
plt.xticks(x_values)
plt.ylabel("Time (s)")
plt.title("Time to compute barcodes against number of interpolated points")
plt.show()
```

```python
four_extra_points = interpolate(data_3kzn, 4)
tic = time.time()
ripser_plot(four_extra_points, title = 'Persistence Diagram for 3kzn chain with four extra points')
toc = time.time()
times[4] = toc - tic
```
